```python
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR= Q3 - Q1
outliers = ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)
```

```python
df_clean = df[~outliers]
df_clean
```
